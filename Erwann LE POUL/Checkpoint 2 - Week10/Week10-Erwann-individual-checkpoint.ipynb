```python
# Load and inspect step count data with hourly time intervals.
# Loading hourly datasets.
df_hour = pd.read_csv("hourlySteps_merged.csv")
```

## Determining percentile for different thresholds to use
We will exclude all zeros from the dataset to determine the threshold, as this will provide a more accurate percentile calculation. Including zeros would result in a percentile value of zero for each case.

Beginning of work 1st October 2024, End of work 1st of October 2024

```python
# Filter out rows where StepTotal is 0.
filtered_steps = df_hour[df_hour['StepTotal'] > 0]['StepTotal']

# Calculate the required percentiles.
percentiles = filtered_steps.quantile([0.25, 0.31, 0.37])

# Print the results.
print(percentiles)
```

The output of the code represents the calculated percentiles for the "StepTotal" column, excluding any rows where the step count is zero. Here's what each value means:

- 0.25: This means that 25% of the non-zero step totals in the dataset are 103 steps or fewer.
- 0.31: This indicates that 31% of the non-zero step totals are 141 steps or fewer.
- 0.37: This shows that 37% of the non-zero step totals are 180 steps or fewer.

## Preparing the data for analysis
In this section, we are analysing the activity levels of a specific individual (index 9), identified by their unique ID from the dataset. The purpose is to determine the number of hours per day this individual is "active" based on different step count thresholds. We will look into the following:

- Filter the Data: Select the data for the individual located at index 9 in the dataset.
- Group by Day: Convert the hourly data into daily data to track the number of active hours per day.
- Apply Thresholds: Define different step count thresholds (103, 141, 180, and 235 steps) and evaluate how many hours per day the individual meets or exceeds these thresholds.
- Store Results: For each day, we count the number of hours the individual was active based on each threshold and store the results in a DataFrame.
- Output the Data: Print or save the resulting data for further analysis and visualizations.

Before analysing the data, our assumption is that the individual at index 9 will exhibit varying levels of activity across different days, with more consistency on weekdays compared to weekends. We predict that the individual will be considered "active" for more hours per day, the lower the threshold. In addition, we expect a correlation between active hours and typical daily routines, with higher activity during workdays and potentially less activity on weekends.

Beginning of work 1st October 2024, End of work 1st of October 2024

```python
# Since we are interested in analysing the activity of a specific individual,
# We first retrieve the unique IDs of all individuals in the dataset.
# We select the individual at index 9.
unique_hour_ids = df_hour["Id"].unique()
individual_id = unique_hour_ids[9]

# Filter the data to extract only the data corresponding to this individual.
individual_data = df_hour[df_hour['Id'] == individual_id].copy()

# To group the data by day, we need to convert the 'ActivityHour' column, which stores date and time, into a datetime object.
individual_data['ActivityHour'] = pd.to_datetime(individual_data['ActivityHour'])

# Next, we create a new column 'Day' that extracts the date from the 'ActivityHour'. This allows us to aggregate the data by day.
individual_data['Day'] = individual_data['ActivityHour'].dt.date

# Define thresholds to evaluate how many hours per day the individual is active based on each threshold.
thresholds = [103, 141, 180, 235]

# Create a DataFrame to store the results, 
# The rows represent the unique days, and the columns represent the different thresholds.
active_hours_per_day = pd.DataFrame(index=individual_data['Day'].unique(), columns=thresholds)

# For each threshold, we calculate how many hours per day the individual is "active" 
# by checking if the step count in each hour exceeds the threshold.
for threshold in thresholds:
    # Add a new column 'Active' to indicate whether the individual was active 
    # in each hour (i.e., if the step count is greater than or equal to the threshold).
    individual_data['Active'] = individual_data['StepTotal'] >= threshold
    
    # Group the data by day and sum the 'Active' values to get the total number 
    # of active hours for each day at the current threshold.
    active_hours_per_day[threshold] = individual_data.groupby('Day')['Active'].sum()

# Finally, we print the resulting DataFrame, which shows the number of active hours per day for each of the thresholds.
print(active_hours_per_day)
```

## Conclusion on Step Count Threshold Variant:
In this report, we analysed the impact of different step count thresholds (103, 141, 180, and 235 steps per hour) from the percentile of the dataset. Our goal was to identify which threshold provides a more realistic and appropriate measure of activity based on the individual's daily patterns.

**Key Observations:**  

- Lower Thresholds (103 and 141 steps): These thresholds classified more hours per day as "active," resulting in a higher number of active hours. They are useful for identifying general movement, including low-intensity activities such as walking or light physical activity. However, these thresholds might overestimate the number of active hours since even minimal movement qualifies as "active."
- Higher Thresholds (180 and 235 steps): These thresholds significantly reduced the number of active hours, especially at the 235-step threshold, where a greater amount of movement was required to qualify as "active." While these thresholds may more accurately capture periods of moderate to high-intensity activity, they could underestimate activity by not accounting for sustained low-intensity movement.
- Daily Activity Patterns: Across all thresholds, there were noticeable variations in activity, particularly on weekends or rest days where active hours dropped significantly. The higher thresholds (180 and 235 steps) tended to highlight these drops more dramatically.
The heatmap and bar plots demonstrated that lower thresholds provided more consistent estimates of daily activity, while higher thresholds captured only the most active periods.

**Recommendation:**  

Based on the analysis, the most appropriate threshold depends on the context of the activity measurement:

- For general activity monitoring, where the goal is to track any form of movement (including light activity), a lower threshold like 103 or 141 steps is more appropriate. These thresholds provide a broader view of the individualâ€™s overall activity levels, making them suitable for tracking daily health or fitness goals.
- For more focused, moderate-to-high intensity activity tracking, where the aim is to capture periods of significant movement or exercise, a higher threshold like 180 or 235 steps would be more suitable. These thresholds ensure that only sustained, vigorous activity is counted as "active."

In conclusion, it depends on the final definition of inactivity and activity. For our case a 180-step threshold strikes a good balance between capturing general activity without being too lenient, making it a practical recommendation for most use cases. It avoids overestimating activity while still being sensitive to light movement that contributes to overall health.
